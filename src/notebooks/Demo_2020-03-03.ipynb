# How to _plot_ data, _save_ the plot and _load_ an image

```python
import numpy as np
import matplotlib.pyplot as plt

import os

plt.style.use('classic')

%matplotlib inline
```

## Create Dummy Data

```python
x = np.arange(10)
y = x**2 + 10
```

## Make Plot and Save it to File-system

```python
fig, ax = plt.subplots()
plt.plot(x,y)
plt.show()
fig.savefig('myplot.png', dpi=300)
```

## Check if the plot was saved

```python
os.listdir()
```

## Loading an Existing Image

We load in the image we just saved _myplot.png_ using `plt.imread()` and show it using `plt.imshow()`.

```python
image = plt.imread('myplot.png')
plt.imshow(image)
plt.show()
```
